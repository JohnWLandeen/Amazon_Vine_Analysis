```python
import numpy as np
import pandas as pd
```

```python
vine = pd.read_csv('data/vine_table.csv')
```

```python
vine.head()
```

```python
# Filtering for total_votes greater than 20.
greater_total_vine = vine[vine['total_votes']>=20]
greater_total_vine
```

```python
# Filtering for all rows where the helpful votes divided by the total votes is greater than 50%.
greater_vote_vine = greater_total_vine[greater_total_vine['helpful_votes'] / greater_total_vine['total_votes'] >= 0.5]
greater_vote_vine
```

```python
# Filtering for all rows where the review was written as part of the paid program.
paid_vine = greater_vote_vine[greater_total_vine['vine']=='Y']
paid_vine
```

```python
# Filtering for all rows where the review was not written as part of the paid program.
unpaid_vine = greater_vote_vine[greater_total_vine['vine']=='N']
unpaid_vine
```

```python
# Total count of reviews in paid_vine
paid_vine['review_id'].count()
```

```python
# Total count of reviews in unpaid_vine
unpaid_vine['review_id'].count()
```

```python
# paid_vine count of 5 star reviews.
paid_vine[paid_vine['star_rating']== 5.0].count()
```

```python
# unpaid_vine count of 5 star reviews.
unpaid_vine[unpaid_vine['star_rating']== 5.0].count()
```

```python
# paid_vine percentages of star reviews.
paid_vine['star_rating'].value_counts(normalize=True) * 100
```

```python
# unpaid_vine percentages of star reviews.
unpaid_vine['star_rating'].value_counts(normalize=True) * 100
```
